3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
def experience(input_string):
    # Если строка пустая или не содержит нужной нам информации, возвращаем nan
    if input_string == np.nan or input_string == 'Не указано' or \
                                    type(input_string) is float:
        res = np.nan
    else:
        # Выделяем содержательную часть строки
        processed_string = input_string.split('  ')[0]
        # Убираем "Опыт работы", оставляем только значимое
        words = processed_string.split()[2:]
        # Обрабатываем полученный списк из 2 или 4 элементов
        if len(words) == 4:
            res = int(words[0]) * 12 + int(words[2])
        elif 'месяц' in words[1]:
            res = int(words[0])
        else:
            res = int(words[0]) * 12
    return res

data['Опыт работы (месяц)'] = data['Опыт работы'].apply(experience)
```

```python
million_cities = [
    'Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань', 'Челябинск',
    'Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 
    'Волгоград' 
    ]

def city_reloc_trip(input_string):
    reloc = False
    trip = False
    info_strings = input_string.split(' , ')
    if info_strings[0] in million_cities:
        city = 'город-миллионник'
    elif info_strings[0] in ['Москва', 'Санкт-Петербург']:
        city = info_strings[0]
    else:
        city = 'другие'
    # Информацию о метро (2 элемент списка не о переезде) убираем
    if 'перее' not in info_strings[1]:
        info_strings.pop(1)
    # Далее достаточно проверить отсутствие отрицания в начале строки
    if info_strings[1].split()[0] != 'не':
        reloc = True
    # Та же логика, а также проверка наличия данных по командировкам
    if len(info_strings) > 2 and info_strings[2].split()[0] != 'не':
        trip = True
    return pd.Series([city, reloc, trip])

data[['Город', 'Готовность к переезду', 'Готовность к командировкам']
     ] = data['Город, переезд, командировки'].apply(city_reloc_trip)
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
employment = set([])
schedule = set([])

# Функция, собирающая уникальные значения признака
def join_func(input_str, to_set):
    items = str(input_str).split(', ')
    to_set.update(items)
 
# Функция, реализующая one hot encoding заданного признака   
def one_hot_func(input_str, input_item):
    items = str(input_str).split(', ')
    if input_item in items:
        return True
    else:
        return False

# Получаем уникальные значения признаков    
for item in data['Занятость'].tolist():
    join_func(item, employment)
for item in data['График'].tolist():
    join_func(item, schedule)

# Чтобы не делать дважды одно и то же, объединяем значения признаков   
united_data = data['Занятость'] + ', ' + data['График']

# По каждому из уникальных значений признаков делаем one hot encoding
# Сначала уникальные значения объединяем, чтобы не повторять дважды
for item in employment.union(schedule):
    data[item] = united_data.apply(one_hot_func, input_item=item)
```

```python
mask1 = (data['проектная работа'] == True)
mask2 = (data['волонтерство'] == True)
mask3 = (data['вахтовый метод'] == True)
mask4 = (data['гибкий график'] == True)

print(data.loc[mask1 & mask2, 'Пол'].count())
print(data.loc[mask3 & mask4, 'Пол'].count())
```

```python
# Удаляем ненужные признаки
data.drop(columns=['Занятость', 'График'], inplace=True)
```

4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
money = pd.read_csv('ExchangeRates.csv')
money['date'] = pd.to_datetime(money['date'], dayfirst=True).dt.date

data['Обновление резюме'] = pd.to_datetime(data['Обновление резюме'],
                                           dayfirst=True).dt.date

def split_salary(x):
    return pd.Series([int(x.split()[0]), x.split()[1]])

data[['ЗП цифра', 'ЗП валюта']] = data['ЗП'].apply(split_salary)
```

```python
currency_dict = {'AZN': 'AZN', 'EUR': 'EUR', 'KGS': 'KGS', 'KZT': 'KZT',
                 'USD': 'USD', 'бел.руб.': 'BYN', 'грн.': 'UAH',
                 'сум': 'UZS' }

def convert_iso(x):
    if x in currency_dict.keys():
        return currency_dict[x]
    else:
        return x
    
data['ЗП валюта ISO'] = data['ЗП валюта'].apply(convert_iso)

data = data.merge(money, left_on=['Обновление резюме', 'ЗП валюта ISO'],
                  right_on=['date', 'currency'], how='left')
```

```python
data.loc[data['ЗП валюта'] == 'руб.', 'proportion'] = 1
data.loc[data['ЗП валюта'] == 'руб.', 'close'] = 1

data['ЗП (руб)'] = (data['ЗП цифра'] * data['close'])/data['proportion']
data['ЗП (руб)'] = data['ЗП (руб)'].astype(int)
data['ЗП (руб)'].median()
```

```python
# Удаляем ненужные признаки
data.drop(columns=['ЗП', 'ЗП цифра', 'ЗП валюта', 'ЗП валюта ISO', 'currency',
                   'per', 'date', 'time', 'close', 'vol', 'proportion'
                   ], inplace=True)
data.info()
```

Правая часть распределения более длинная, то есть распределение правостороннее.

```python
def outliers_z_score_mod(data, feature, left=3, right=3, log_scale=False):
    # Получаем признак, в котором будем обрабатывать выбросы
    if log_scale:
        x = np.log(data[feature] + 1)
    else:
        x = data[feature]
    # Вычисляем среднее и стандартное отклонение
    mu = x.mean()
    sigma = x.std()
    # Вычисляем границы выбросов
    lower_bound = mu - left * sigma
    upper_bound = mu + right * sigma
    # Получаем выбросы и очищенне данные
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x >= lower_bound) & (x <= upper_bound)]
    return outliers, cleaned
```

```python
to_clean, result = outliers_z_score_mod(data, 'Возраст', left=3, right=4, 
                                        log_scale=True)
to_clean
```
